```python
#importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
#declaring empty lists
mc=[] #empty list for college name
mo=[] #empty list for ownership
mr=[] #empty list for rank
mco=[] #empty list for courses and fees
fl=[] #empty list for facility links
#Buildind a function to get the required data from the content
def repeat(soup):
    ownerships=soup.find_all('strong', class_='strong_ownership')
    colleges=soup.find_all('h3', class_='college_name d-md-none')
    ranking=soup.find_all('div', class_='align-center')
    courses=soup.find_all('div',class_='snippet_block border_style')
    links=soup.find_all('ul',class_='links_list')
    for college in colleges:
        mc.append(college.text)
    for i in range(len(ownerships)):
        if i%2==0:
            mo.append(ownerships[i].text)
        else:
            continue
    for rank in ranking:
        mr.append(rank.text)
    for course in courses:
        mco.append(course.text)
    for i in range(len(links)):
        if i%2==0:
            facility_link=links[i].find('a',string="Facilities")
            fl.append(facility_link['href'])
        else:
            continue
```

```python
#reading the contents of first 4 webpages to get data of top 200 colleges
for i in range(1,5):
    url=f"https://engineering.careers360.com/colleges/ranking?page={i}&sort_by=1"
    r=r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    repeat(soup)
```

```python
#the college at 179th position does not have any courses in the website
#to maintain the order of remaining columns pop 179th element
mc.pop(179)
mo.pop(179)
mr.pop(179)
fl.pop(179)
```

```python
#getting primary courses and fees from the list mco as the list contains both courses and fees
primary_courses=[course.split(')')[0]+')' for course in mco]
pattern = r'₹[\d\.]+ (?:Lakhs?|K)'
extracted_fees = []
for course_str in mco:
    fee = re.findall(pattern, course_str)
    extracted_fees.append(fee)
fees = [row[0] if row else 'N/A' for row in extracted_fees]
```

```python
#every list contains same number of elements
#defining dictionary in a required form
d={"NIRF Rank'23":mr,"College_Name":mc,"Ownership":mo,"Primary_Course":primary_courses,"Fees":fees,"Facilities_link":fl}
print(d)
```

```python
#reading dictionary into model
model=pd.DataFrame(data=d)
print(model)
```
